### Confidence Intervals

Confidence intervals provide a range of values within which the true population parameter (like the mean) is likely to lie, based on a sample from that population. A 95% confidence interval means that if we were to take many random samples and calculate a confidence interval for each, approximately 95% of those intervals would contain the true population mean.

We will estimate a 95% confidence interval for the mean of the 'trip_distance', 'fare_amount', and 'tip_amount' columns.

```python
import numpy as np
from scipy import stats

# Columns for which to calculate confidence intervals
cols_for_ci = ['trip_distance', 'fare_amount', 'tip_amount']

for col in cols_for_ci:
    # Drop missing values for the calculation
    data = df[col].dropna()

    # Calculate the mean
    mean = np.mean(data)

    # Calculate the standard error of the mean (SEM)
    # SEM = standard deviation / sqrt(n)
    sem = stats.sem(data)

    # Calculate the 95% confidence interval
    # For a 95% CI, alpha = 0.05, so alpha/2 = 0.025. We use the percent point function (ppf)
    # of the t-distribution to get the t-score for the desired confidence level.
    # Degrees of freedom (df) = n - 1
    ci = sem * stats.t.ppf((1 + 0.95) / 2., len(data)-1)

    confidence_interval = (mean - ci, mean + ci)

    print(f"95% Confidence Interval for Mean {col}: {confidence_interval}")
```

**Interpretation of the Confidence Intervals:**

After running the code, you will see a range for each of the selected columns. For example, if the output for 'Mean trip_distance' is (5.0, 6.0), it means we are 95% confident that the true average trip distance in the entire population from which this data was sampled is between 5.0 and 6.0 miles.

It's important to remember that the confidence interval is about the population mean, not the individual data points.

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load the dataframe
df = pd.read_parquet('/content/drive/MyDrive/sem5/ds_assignments/yellow_tripdata_2025-06.parquet',engine='pyarrow')

# Columns for which to calculate confidence intervals
cols_for_ci = ['trip_distance', 'fare_amount', 'tip_amount']

for col in cols_for_ci:
    # Drop missing values for the calculation
    data = df[col].dropna()

    # Calculate the mean
    mean = np.mean(data)

    # Calculate the standard error of the mean (SEM)
    # SEM = standard deviation / sqrt(n)
    sem = stats.sem(data)

    # Calculate the 95% confidence interval
    # For a 95% CI, alpha = 0.05, so alpha/2 = 0.025. We use the percent point function (ppf)
    # of the t-distribution to get the t-score for the desired confidence level.
    # Degrees of freedom (df) = n - 1
    ci = sem * stats.t.ppf((1 + 0.95) / 2., len(data)-1)

    confidence_interval = (mean - ci, mean + ci)

    print(f"95% Confidence Interval for Mean {col}: {confidence_interval}")
```

### Hypothesis Testing - One-Sample t-test

A one-sample t-test is used to determine whether the mean of a single sample is significantly different from a known or hypothesized population mean.

In this case, we want to test the following hypotheses:

*   **Null Hypothesis (H0):** The average tip amount in the population is equal to $2.
*   **Alternative Hypothesis (H1):** The average tip amount in the population is different from $2.

We will use a significance level (alpha) of 0.05. If the p-value from the test is less than 0.05, we will reject the null hypothesis.

```python
from scipy import stats

# Hypothesis test: One-sample t-test for mean tip amount
# H0: mean tip amount = 2
# H1: mean tip amount != 2

# Drop missing values from the 'tip_amount' column
tip_amounts = df['tip_amount'].dropna()

# Perform the one-sample t-test
# The second argument is the hypothesized population mean (mu)
t_statistic, p_value = stats.ttest_1samp(tip_amounts, 2)

print(f"One-Sample t-test for Mean Tip Amount vs $2:")
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print(f"Since the p-value ({p_value:.4f}) is less than alpha ({alpha}), we reject the null hypothesis.")
    print("Conclusion: The average tip amount is significantly different from $2.")
else:
    print(f"Since the p-value ({p_value:.4f}) is greater than or equal to alpha ({alpha}), we fail to reject the null hypothesis.")
    print("Conclusion: There is not enough evidence to conclude that the average tip amount is significantly different from $2.")
```

**Interpretation of the One-Sample t-test Results:**

Look at the P-value from the output.

*   If the P-value is very small (typically less than 0.05), it means that the observed sample mean is very unlikely to occur if the true population mean was $2. In this case, you would reject the null hypothesis and conclude that the average tip amount is significantly different from $2.
*   If the P-value is large (typically greater than or equal to 0.05), it means that the observed sample mean is reasonably likely to occur if the true population mean was $2. In this case, you would fail to reject the null hypothesis and conclude that there is not enough evidence to say the average tip amount is significantly different from $2.

The t-statistic measures the difference between the sample mean and the hypothesized population mean in units of the standard error. A larger absolute t-statistic indicates a greater difference.

### Hypothesis Testing - Two-Sample t-test

A two-sample t-test is used to compare the means of two independent groups to determine if there is a significant difference between them.

In this case, we will compare the average fare amount for trips paid with Payment Type 1 (likely credit card) and Payment Type 2 (likely cash).

*   **Null Hypothesis (H0):** $\mu_{Payment Type 1} = \mu_{Payment Type 2}$ (The average fare amount is the same for both payment types).
*   **Alternative Hypothesis (H1):** $\mu_{Payment Type 1} \neq \mu_{Payment Type 2}$ (The average fare amount is different for the two payment types).

We will use a significance level (alpha) of 0.05.

```python
from scipy import stats

# Hypothesis test: Two-sample t-test for average fare amount between two payment types

# Filter data for the two payment types (assuming 1 and 2 based on bar chart)
fare_type1 = df[df['payment_type'] == 1]['fare_amount'].dropna()
fare_type2 = df[df['payment_type'] == 2]['fare_amount'].dropna()

# Perform the independent two-sample t-test
# We assume unequal variances (Welch's t-test) which is often a safe assumption
t_statistic_2sample, p_value_2sample = stats.ttest_ind(fare_type1, fare_type2, equal_var=False)

print(f"Two-Sample t-test for Average Fare Amount (Payment Type 1 vs 2):")
print(f"T-statistic: {t_statistic_2sample}")
print(f"P-value: {p_value_2sample}")

# Interpret the results
alpha = 0.05
if p_value_2sample < alpha:
    print(f"Since the p-value ({p_value_2sample:.4f}) is less than alpha ({alpha}), we reject the null hypothesis.")
    print("Conclusion: The average fare amount for Payment Type 1 is significantly different from Payment Type 2.")
else:
    print(f"Since the p-value ({p_value_2sample:.4f}) is greater than or equal to alpha ({alpha}), we fail to reject the null hypothesis.")
    print("Conclusion: There is not enough evidence to conclude that the average fare amount for Payment Type 1 is significantly different from Payment Type 2.")
```

**Interpretation of the Two-Sample t-test Results:**

Similar to the one-sample t-test, the P-value is key to the interpretation.

*   If the P-value is less than 0.05, you would reject the null hypothesis and conclude that there is a statistically significant difference in the average fare amount between the two payment types.
*   If the P-value is 0.05 or greater, you would fail to reject the null hypothesis, meaning there is not enough statistical evidence to conclude that the average fare amounts are different between the two groups.

The t-statistic indicates the magnitude of the difference between the two group means relative to the variability within the groups.

### Hypothesis Testing - Chi-square Test of Independence

The Chi-square Test of Independence is used to determine whether there is a significant association between two categorical variables. It tests if the observed frequencies in a contingency table are significantly different from the frequencies that would be expected if the variables were independent.

In this case, we want to test if 'Payment type' and 'RateCodeID' are independent.

*   **Null Hypothesis (H0):** 'Payment type' and 'RateCodeID' are independent. (There is no association between payment method and the rate code of the trip).
*   **Alternative Hypothesis (H1):** 'Payment type' and 'RateCodeID' are not independent. (There is an association between payment method and the rate code of the trip).

We will use a significance level (alpha) of 0.05. If the p-value from the test is less than 0.05, we will reject the null hypothesis and conclude that the variables are not independent.

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Create a contingency table of 'Payment type' and 'RateCodeID'
# We'll drop any rows with missing values in these columns for the test
contingency_table = pd.crosstab(df['payment_type'], df['RatecodeID'])

# Perform the Chi-square test of independence
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square Test of Independence between Payment type and RateCodeID:")
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
# print(f"Expected frequencies table:\n{expected}") # Uncomment to see the expected frequencies

# Interpret the results
alpha = 0.05
if p < alpha:
    print(f"Since the p-value ({p:.4f}) is less than alpha ({alpha}), we reject the null hypothesis.")
    print("Conclusion: Payment type and RateCodeID are not independent.There is a significant association between them.")
else:
    print(f"Since the p-value ({p:.4f}) is greater than or equal to alpha ({alpha}), we fail to reject the null hypothesis.")
    print("Conclusion: There is not enough evidence to conclude that Payment type and RateCodeID are not independent.")
```

# **Interpretation of the Chi-square Test of Independence Results:**

*   **Chi-square statistic:** This is a measure of the difference between the observed frequencies in your data and the frequencies that would be expected if the two variables were truly independent. A larger Chi-square statistic suggests a greater difference between observed and expected frequencies, indicating a potential association.
*   **P-value:** This is the probability of observing a Chi-square statistic as extreme as, or more extreme than, the one calculated from your data, assuming the null hypothesis (independence) is true.
    *   If the P-value is less than your chosen significance level (alpha, which is 0.05), you have strong evidence against the null hypothesis. You would reject the null hypothesis and conclude that the two variables are **not independent**; there is a statistically significant relationship between them.
    *   If the P-value is greater than or equal to alpha, you do not have enough evidence to reject the null hypothesis. You would fail to reject the null hypothesis and conclude that there is **not enough evidence to say that the two variables are not independent**. This doesn't necessarily mean they *are* independent, just that your data doesn't provide enough evidence to claim they are dependent at the chosen significance level.
*   **Degrees of freedom (dof):** This value is determined by the number of categories in each variable in your contingency table. It influences the shape of the Chi-square distribution used to calculate the p-value. The formula is (number of rows - 1) \* (number of columns - 1).
*   **Expected frequencies table:** (If you uncommented the line to print it) This table shows the frequencies that would be expected in each cell of the contingency table if the two variables were perfectly independent, based on the row and column totals. Comparing the observed frequencies (in the `contingency_table` you created) to these expected frequencies helps understand where the differences lie.

In summary, the Chi-square test tells you if the relationship between 'Payment type' and 'RateCodeID' you observe in your data is likely due to a real association in the population or just random chance.

### Correlation Analysis

Correlation analysis helps us understand the strength and direction of the linear relationship between two numerical variables.

*   **Pearson correlation coefficient:** Measures the linear relationship between two continuous variables. It ranges from -1 to +1, where:
    *   +1 indicates a perfect positive linear relationship.
    *   -1 indicates a perfect negative linear relationship.
    *   0 indicates no linear relationship.
*   **Spearman correlation coefficient:** Measures the monotonic relationship between two variables (whether they tend to increase or decrease together, not necessarily at a constant rate). It's useful for non-normally distributed data or when the relationship is not strictly linear. It also ranges from -1 to +1.

We will compute the Pearson correlation between 'Trip distance' and 'Fare amount', and between 'Fare amount' and 'Tip amount'. We will also create a correlation matrix heatmap to visualize the relationships between several numerical columns.

```python
# Select columns for correlation analysis
correlation_cols = ['trip_distance', 'fare_amount', 'tip_amount', 'total_amount', 'extra']

# Compute Pearson correlation
pearson_corr = df[correlation_cols].corr(method='pearson')

# Compute Spearman correlation (optional, depending on data distribution)
spearman_corr = df[correlation_cols].corr(method='spearman')

print("Pearson Correlation Matrix:")
display(pearson_corr)

# print("\nSpearman Correlation Matrix:")
# display(spearman_corr) # Uncomment to see Spearman correlation

# Create a correlation matrix heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Pearson Correlation Heatmap')
plt.show()
```
